# OECD Aid statistics

Downloaded all aids by recipients from http://stats.oecd.org/qwids/  - results.csv

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline  
```

```python
OECD=pd.read_csv('results.csv')
OECD.head(3) # columns are a bit fucked...
```

```python
OECD=pd.read_csv('results.csv', skiprows=1)
OECD.head(3) 
OECD.index=OECD['Unnamed: 0']
OECD=OECD.transpose()
OECD=OECD.drop(OECD.index[[0,1]])
OECD.head()
```

## Total Amount of Aid per Recipient Country

```python
OECD[['Niger','Yemen','Somalia']].convert_objects(convert_numeric=True).plot()
plt.ylabel('Aid [USD millions]')
```
