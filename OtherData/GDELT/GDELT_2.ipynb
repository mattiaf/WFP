```python
import numpy as np
%matplotlib inline  
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# Build a mapper to convert from event code to event description
Event_code = pd.read_csv('CAMEO.eventcodes.txt', sep='\t')
Event_code.head()
mapper = {}
for i in Event_code.index:
    mapper[Event_code.ix[i].CAMEOEVENTCODE]=Event_code.ix[i].EVENTDESCRIPTION
Event_code.head()
```

```python
# read Yemen data
df_YEM=pd.read_csv('YEM.csv') 

# transform in a table with months as columns and events as rows
df_YEM=df_YEM.pivot(columns='MonthYear',index='EventCode',values='N_events') 
df_YEM=df_YEM.fillna(0)

# create column with event description
df_YEM['EventDescription']=df_YEM.index
df_YEM.EventDescription=df_YEM.EventDescription.map(mapper)

# orders column
df_YEM=df_YEM.reset_index()
df_YEM.columns.name = 'index'
columnordered=['EventCode','EventDescription']
list2=list(df_YEM.columns[1:-1])
columnordered.extend(list2)
df_YEM=df_YEM[columnordered]
```
