# USDA projections on GDP and population

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline  
```

```python
pd.read_excel('data/ProjectedPopulationValues.xls')
```

```python
#let's make that usable...
Pop=pd.read_excel('data/ProjectedPopulationValues.xls',skiprows=12)
Pop.head() # Pop['Unnamed: 3']  are countries, ones with NaNs are regions
Pop = Pop[Pop['Unnamed: 3'].isnull() == False]
```

```python
Pop.index= Pop['Unnamed: 3'] 
Pop=Pop.drop(['Unnamed: 0','Unnamed: 1','Unnamed: 2','Unnamed: 3'],1)
Pop=Pop[range(2010,2031,1)].transpose().head()
Pop.head()
```

```python
Pop.index=pd.to_numeric(Pop.index)
Pop[['Iraq','Niger','Ukraine']].plot()
plt.title('projected Population 2010-2030')
locs,labels = plt.xticks()

plt.xticks(locs, map(lambda x: "%g" % x, locs))
```

```python
#let's make that usable...
GDP=pd.read_excel('data/ProjectedRealPerCapitaGDPValues.xls',skiprows=12)
GDP.head() # Pop['Unnamed: 3']  are countries, ones with NaNs are regions
GDP = GDP[GDP['Unnamed: 3'].isnull() == False]
GDP.index= GDP['Unnamed: 3'] 
GDP=GDP.drop(['Unnamed: 0','Unnamed: 1','Unnamed: 2','Unnamed: 3'],1)
GDP=GDP[range(2010,2031,1)].transpose().head()
GDP.head()
GDP.index=pd.to_numeric(GDP.index)
GDP[['Iraq','Niger','Ukraine']].plot()
plt.title('projected GDP pro capita 2010-2030')
```
